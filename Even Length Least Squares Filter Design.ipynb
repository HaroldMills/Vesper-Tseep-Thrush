# Preliminaries

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
import scipy.signal as signal
from scipy.linalg import toeplitz, hankel, pinv

import dsp_utils

%matplotlib inline
```

# Design filters of length 100 and 101.
The filter of length 100 is designed by our new code, and the filter of length 101 by SciPy.

```python
fs = 22050
fs2 = fs / 2
f0 = 6000
f1 = 10000
bw = 100

bands = np.array([0, f0 - bw, f0, f1, f1 + bw, fs2]) / fs2
desired = np.array([0, 0, 1, 1, 0, 0])
h_100 = dsp_utils.firls(100, bands, desired)
h_101 = dsp_utils.firls(101, bands, desired)
```

# Plot filter frequency responses.

```python
# Get frequency responses.
w, H_100 = signal.freqz(h_100)
_, H_101 = signal.freqz(h_101)
w *= fs / (2 * np.pi)


# Plot frequency responses.

figure = plt.figure(figsize=(15, 4))
plt.title('Filter Frequency Responses')

plt.plot(w, 20 * np.log10(abs(H_100)), 'r')
plt.plot(w, 20 * np.log10(abs(H_101)), 'b')
plt.ylabel('Amplitude (dB)', color='b')
plt.xlabel('Frequency (Hz)')

plt.grid()
plt.show()
```
